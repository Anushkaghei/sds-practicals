Binomial Distribution

```python
import matplotlib.pyplot as plt
from math import factorial as fact

def combination(n,r):
    return(fact(n)/(fact(r)*fact(n-r)))

def binomial(n,success_prob):
    prob=[]
    for x in range(n+1):#no. of heads
        p=combination(n,x)*pow(success_prob,x)*pow(1-success_prob,n-x)
        prob.append(p)
    plt.bar(range(n+1),prob,width=1)
    
binomial(10,0.1)
```

Poisson Distribution

```python
import matplotlib.pyplot as plt
import math

def poisson(l):
    prob=[]
    p_x=1
    x=0
    
    while(p_x>0.00000000001):
        p_x=math.exp(-1)*pow(l,x)/math.factorial(x)
        prob.append(p_x)
        x=x+1
    plt.bar(range(x),prob)
    plt.show()
    
poisson(2)
poisson(3)
poisson(5)
poisson(10)
```

Normal distribution

```python
import matplotlib.pyplot as plt
import math

def normal(mu,sigma):
    prob=[]
    for x in range(-4*sigma+mu,4*sigma+mu):#no. of heads
        p_x=(1/(sigma*math.sqrt(2*math.pi)))*math.exp(-1*pow(((x-mu)/sigma),2)/2)
        prob.append(p_x)
        
    plt.plot(range(-4*sigma+mu,4*sigma+mu),prob)
    plt.show()
    
```

```python
normal(30,10)
```
