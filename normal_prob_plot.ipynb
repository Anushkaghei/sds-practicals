```python
from scipy.stats import zscore
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
#its just an example npp of original values v/s theortical values(z scores)
def npp(data):
    data = sorted(data)
    p = [(data.index(i)-0.5)/len(data) for i in data]
    z = zscore(p)

    plt.scatter(data, z)
    plt.ylabel('Z values')
    plt.xlabel('theortical values')
    plt.show()
    plt.plot(data,  z,'ro',data,  data)
    plt.show()
    
#n datapoints
n = 100       
data = np.random.randn(n) #The numpy. random. randn() 
#function creates an array of specified shape and fills it with 
#random values as per standard normal distribution.
npp(data)
```

```python

X1 = [3.01, 3.35, 4.79, 5.96, 7.89]

X2 = [3.89, 4.75, 4.75, 5.20, 5.78, 5.80, 6.33, 7.21, 7.90]

X3 = [108.047, 109.249, 103.385, 112.454, 95.780,  122.734, 109.842 , 120.858, 
    98.604, 98.122,  95.971, 98.173,   87.437 , 91.884, 92.193,  83.882]
def npp1(data1):
    p =[]
    t =[]
    data1 = np.sort(np.array(data1))
    p = [(i - 0.5)/len(data1) for i in range(1, len(data1)+1)]
    t = [norm.ppf(i, np.mean(data1), np.std(data1)) for i in p]
    plt.plot(data1, t, 'ro', data1, data1)
    plt.show()

npp1(X1)
npp1(X2)
npp1(X3)
```
