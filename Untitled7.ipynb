```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv("height-weight.csv")

X = data['height'].iloc[:50]
y = data['weight'].iloc[:50]
n = len(X)

sum_of_xi_yi = 0

xbar = np.mean(X)
ybar = np.mean(y)

xi_sq = 0
yi_sq = 0

for i in range(n):
    sum_of_xi_yi += X[i]*y[i]
    xi_sq += X[i]*X[i]
    yi_sq += y[i]*y[i]

B1 = (sum_of_xi_yi - n*xbar*ybar)/(xi_sq - n*xbar*xbar)
B0 = ybar - B1*xbar
r = (sum_of_xi_yi - n*xbar*ybar)/((xi_sq - n*xbar*xbar)*(yi_sq - n*ybar*ybar))**0.5


print(f"Beta 1 is {B1}")
print(f"Beta 0 is {B0}")
print(f"R is {r}")

plt.scatter(X,y)
Y_hat = []
for i in range(n):
    Y_hat.append(B0 + B1*X[i])
plt.plot(X,Y_hat)

plt.show()
```

```python
x_inp = int(input("Enter height: "))
y_pred = B0 + B1*x_inp
print(f"Predicted weight for height {x_inp} is {y_pred}")
```

```python
residual=[]
for i in range(n):
  residual.append(y[i]-Y_hat[i])
```

```python
plt.scatter(y,residual,color='g')
plt.axhline(y=0,color='r')
```

```python
#power of test
from math import sqrt
from scipy.stats import norm
import matplotlib.pyplot as plt

n=30
sigma =10
alpha=0.05

#H0 :mu>=50
#H1 :mu<50 //left tailed test; Power : area to left

mu0=50

#moving in direction of alternative: power increases
z_nd= norm.ppf(0.05)

se=sigma/sqrt(n)

critical_xbar= z_nd *se +mu0
power=[]

for i in range(40,mu0):
  z_ad=(critical_xbar-i)/se
  power.append(norm.cdf(z_ad)) #for area to left use cdf for right 1- cdf
```

```python
n_list=[10,30,100,500]

z_nd= norm.ppf(0.05)

for sample_size in n_list:
  se=sigma/sqrt(sample_size)
  critical_xbar =z_nd *se + mu0
  power=[]
  for i in range(40, mu0):
    z_ad =(critical_xbar-i)/se
    power.append(norm.cdf(z_ad))
  label_value = "n= "+str(sample_size)
  plt.plot(range(40,mu0),power,label=label_value)
  plt.legend(loc="lower left")

plt.show()
```

alpha value

```python
alpha_list = [0.08,0.1,0.56,0.01,0.91]
for sample_size in alpha_list:
    z_nd = norm.ppf(0.05)
    se = sigma/np.sqrt(sample_size)
    cx_bar = mu0 + z_nd*se
    power = []
    for i in range(40,mu0):
        z_ad = (cx_bar - i)/se
        power.append(norm.cdf(z_ad))
    plt.plot(range(40,mu0),power,label = 'Alpha values: '+str(sample_size))
plt.legend()
plt.show()
```

```python
sig_list = [10,20,30,49,25,35]
for sample_size in sig_list:
    z_nd = norm.ppf(0.05)
    se = sigma/np.sqrt(sample_size)
    cx_bar = mu0 + z_nd*se
    pow = []
    for i in range(40,mu0):
        z_ad = (cx_bar - i)/se
        power.append(norm.cdf(z_ad))
    plt.plot(range(40,mu0),power,label = 'Sigma values : '+str(sample_size))
plt.legend()
plt.show()
```
