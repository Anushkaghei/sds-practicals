create a dataframe 

```python
import pandas as pd

df=pd.DataFrame({'date':['2021-01-01','2021-01-02','2021-01-03','2021-01-04','2021-01-01','2021-01-02','2021-01-03','2021-01-04'],
                'fruit':['apple', 'apple','apple','apple','mango','mango','mango','mango',],
                'price':[0.80,None,None,1.20,None,2.10,2.00,1.80]})

df['date']=pd.to_datetime(df['date'])
df
```

```python
df.isna()
```

```python
df.isna().sum() # sum nan values in one column 
```

```python
df[df.isna().any(axis=1)]
```

```python
df[df['price'].isna()] #other way of displaying in a particular column how many nan values are there 
```

to fill them missing values with fillna( value which we want to fill , where-ever it is true fill it w value ) method

```python
df['price'].fillna(value=0.85,inplace=True)
df
```

finding mean and filling value 

```python
df['price'].fillna(value=df.price.mean(),inplace=True)
df

#it gets fillied with 1.58
```

finding mean and filling missing values 

```python
df['price'].fillna(value=df.price.median(),inplace=True)
df

#it gets filled with 1.8
```

above ways must not be done as price may be too high for others 

make two different categories and then find mean and meadian and fill them Nan values 
we make use of groupby attribute 

```python
df['price'].fillna(df.groupby('fruit')['price'].transform('mean'),inplace=True)
df
```

```python
df['price'].fillna(df.groupby('fruit')['price'].transform('median'),inplace=True)
df
```

```python
df.groupby('fruit')['price'].transform('mean')
```

```python
df.groupby('fruit')['price'].transform('median')
```

limit =1 , only one Nan value gets filled with whatever value we give 

do grouping first and then do fflil

```python
df['price']=df.groupby('fruit')['price'].ffill()
df
```

```python
df['price']=df.groupby('fruit')['price'].ffill(limit=1)
df
```

just using ffill

```python
df['price'].fillna(method='ffill',inplace=True)
df
```

```python
df['price'].fillna(method='bfill',inplace=True)
df
```

```python
df['price']=df.groupby('fruit')['price'].bfill()
df
```

```python
df['price']=df.groupby('fruit')['price'].ffill(limit=1)
df
```

use bfill and ffill both in one 

```python
df['price']=df.groupby('fruit')['price'].ffill().bfill()
df
```

```python
df['price']=df.groupby('fruit')['price'].bfill().ffill()
df
```

when we use interpolate , bw two valid values it sees how many values have Nan value , takes average(last value-first value/interval) of those two values and fills the Nan space 

that average value is added to first value and the first Nan space is filled , average value is added to first Nan space value and the second Nan space is filled 

```python
df['price'].interpolate(method='linear',inplace=True)
df
```
