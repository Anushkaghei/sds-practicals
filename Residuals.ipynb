```python
% matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np
import pandas as pd
import statistics

#x vs y

df = pd.read_csv('Residuals.csv')
#x = Wind Speed
#y = Current
plt.scatter(df['Wind Speed'], df.Current, color = 'red')
b1, b0 = np.polyfit(df['Wind Speed'], df.Current, 1)
y_p = b0 + b1 * np.array(df['Wind Speed'])
plt.plot(df['Wind Speed'], y_p)
plt.show()

residuals = np.array(df.Current - y_p)
plt.scatter(y_p, residuals, color='red')
plt.plot([1,4], [0,0])
plt.show()
```

```python
#ln x vs y
plt.scatter(np.log(df['Wind Speed']), df.Current, color = 'red')
b1, b0 = np.polyfit(np.log(df['Wind Speed']), df.Current, 1)
y_p = b0 + b1 * np.array(np.log(df['Wind Speed']))
plt.plot(np.log(df['Wind Speed']), y_p)
plt.show()

residuals = np.array(df.Current - y_p)
plt.scatter(y_p, residuals, color='red')
plt.plot([1,4], [0,0])
plt.show()
```

```python
# x vs ln y

plt.scatter(df['Wind Speed'], np.log(df.Current), color = 'red')
b1, b0 = np.polyfit(df['Wind Speed'], np.log(df.Current), 1)
y_p = b0 + b1 * np.array(df['Wind Speed'])
plt.plot(df['Wind Speed'], y_p)
plt.show()

residuals = np.array(np.log(df.Current) - y_p)
plt.scatter(y_p, residuals, color='red')
plt.plot([0,4], [0,0])
plt.show()
```

```python
# x vs sqrt(y)

plt.scatter(df['Wind Speed'], np.sqrt(df.Current), color = 'red')
b1, b0 = np.polyfit(df['Wind Speed'], np.sqrt(df.Current), 1)
y_p = b0 + b1 * np.array(df['Wind Speed'])
plt.plot(df['Wind Speed'], y_p)
plt.show()

residuals = np.array(np.sqrt(df.Current) - y_p)
plt.scatter(y_p, residuals, color='red')
plt.plot([0,4], [0,0])
plt.show()
```
